```python
## Collision Plot
import numpy as np

groupA_haptic=[4,5,10,6,13,19,18,12,7,14,7,8,3,4,3,7,0,1,2,1]
groupA_no_haptic=[6,5,2,8,13,9,8,10,8,13,12,7,6,11,6,8,4,2,1,4]

groupB_no_haptic=[4,9,4,5,15,13,8,12,4,1,3,2,9,4,4,6,8,9,6,6,11,12,11,13]
groupB_haptic=[3,6,5,6,12,8,8,10,5,3,4,2,9,6,6,5,7,11,10,4,13,11,8,8]

groupA_haptic_mean=np.mean(groupA_haptic)
groupA_no_haptic_mean=np.mean(groupA_no_haptic)

groupB_no_haptic_mean=np.mean(groupB_no_haptic)
groupB_haptic_mean=np.mean(groupB_haptic)


print(groupB_haptic_mean)
```

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

 
# set width of bars
barWidth = 0.25
 
# set heights of bars
bars1 = [groupA_haptic_mean, groupB_no_haptic_mean]
bars2 = [groupA_no_haptic_mean, groupB_haptic_mean]

 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

 
# Make the plot
plt.bar(r1, bars1, color='#9be5aa', width=barWidth, edgecolor='black', label='No Haptic Feedback')
plt.bar(r2, bars2, color='#95d0fc', width=barWidth, edgecolor='black', label='With Haptic Feedback')

 
# Add xticks on the middle of the group bars
plt.xlabel('Groups')
plt.xticks([r + barWidth for r in range(len(bars1))], ['Group A', 'Group B'])
plt.ylabel('Average Number of Collisions') 
plt.legend( loc='lower right')
plt.show()
```
